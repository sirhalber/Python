## Numerical Analysis and Viz

Lets analyze the Liters of pure alcohol consumed per capita column:

```python
alcohol['Liters of pure alcohol consumed per capita'].describe()
```

```python
alcohol['Liters of pure alcohol consumed per capita'].mean()
```

```python
alcohol['Liters of pure alcohol consumed per capita'].median()
```

```python
alcohol['Liters of pure alcohol consumed per capita'].plot(kind='box', vert=False, figsize=(14,6))
```

```python
alcohol['Liters of pure alcohol consumed per capita'].plot(kind='density', figsize=(14,6))
```

```python
ax = alcohol['Liters of pure alcohol consumed per capita'].plot(kind='density', figsize=(14,6))
ax.axvline(alcohol['Liters of pure alcohol consumed per capita'].mean(), color='red')
ax.axvline(alcohol['Liters of pure alcohol consumed per capita'].median(), color='green')
```

```python
ax = alcohol['Liters of pure alcohol consumed per capita'].plot(kind='hist', figsize=(14,6))
ax.set_ylabel('Country')
ax.set_xlabel('Alcohol (L)')
```

## Categorical Analysis and Viz

```python
alcohol['Country'].value_counts()
```

```python
alcohol['Country'].value_counts().plot(kind='pie', figsize=(6,6))
```

```python
ax = alcohol['Country'].value_counts().plot(kind='bar', figsize=(14,6))
ax.set_ylabel('Alcohol')
```

```python
alcohol.plot(kind='bar', x='Country', y='Liters of pure alcohol consumed per capita', figsize=(14,6))
```
